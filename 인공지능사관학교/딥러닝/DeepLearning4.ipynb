```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('/content/faults.csv')

df.info()
```

```python
df_dataset = df[['X_Minimum','X_Maximum','Steel_Plate_Thickness','LogOfAreas',
                 'Pastry','Z_Scratch','Bumps']]
print("df_dataset.shape : ", df_dataset.shape)
print(df_dataset.info())
```

```python
#데이터 모양을 줄이자
df_dataset = np.asarray(df_dataset, dtype = 'float32')

df_dataset_x = df_dataset[0:1,:-3] #독립변수
df_dataset_y = df_dataset[0:1,-3:] #종속변수

print(df_dataset_x.shape)
print(df_dataset_x)
print(df_dataset_y.shape)
print(df_dataset_y)
```

```python
RND_MEAN = 0
RND_STD = 1

input_cnt = df_dataset_x.shape[-1]
ouput_cnt = df_dataset_y.shape[-1]
weight = np.random.normal(RND_MEAN,RND_STD,size = [input_cnt,ouput_cnt]) 
bias = np.random.normal(RND_MEAN,RND_STD,size = [ouput_cnt])
```

```python
#값을 확인하는 습관
print(weight)
print(bias)

```

```python
#for i in range(ouput_cnt)
p_1 = np.matmul(df_dataset_x[0],weight[:,0]) + bias[0]
p_2 = np.matmul(df_dataset_x[0],weight[:,1]) + bias[1]
p_3 = np.matmul(df_dataset_x[0],weight[:,2]) + bias[2]

#weight bias = (model)parameter 값이 바뀌는 존재 
#learning rate x y mean = hyperparameter 값이 안바뀌는 존재 
print(p_1,p_2,p_3)

result = np.matmul(df_dataset_x,weight) + bias

print(result)
```

```python
df = pd.read_csv('/content/faults_mini.csv')
df.info()
display(df)
```

```python
df = np.asarray(df,dtype = 'float32')
```

```python
#데이터의 weight 와 bias 수행 
RND_MEAN = 0
RND_STD = 1

input_cnt = df_dataset_x.shape[-1]
ouput_cnt = df_dataset_y.shape[-1]
weight = np.random.normal(RND_MEAN,RND_STD,size = [input_cnt,ouput_cnt]) 
bias = np.random.normal(RND_MEAN,RND_STD,size = [ouput_cnt])

print(weight)

print(bias)
```

```python
#train test split 
mb_size = 2
train_ratio = 0.8

df.shape #10,7
df.shape[0] * train_ratio #학습 데이터의 개수 

mb_step_count = df.shape[0] * train_ratio // mb_size #미니배치 사이즈를 통한 갯수
print(mb_step_count)
```

```python
#train_test 경계 
test_begin_index = mb_size * mb_step_count
print(test_begin_index)

shffle_map = np.arange(df.shape[0])
np.random.shuffle(shffle_map)

print(shffle_map)
```

```python
mb_data_1 = df[shffle_map[0:2]]
print(mb_data_1)

mb_data_2 = df[shffle_map[2:4]]
print(mb_data_2)

mb_data_3 = df[shffle_map[4:6]]
print(mb_data_3)

mb_data_4 = df[shffle_map[6:8]]
print(mb_data_4)
```

```python
mb_data_1_x = mb_data_1[:,:-3]
mb_data_1_y = mb_data_1[:,-3:]

print(mb_data_1_x)
print(mb_data_1_y)
mb_data_2_x = mb_data_2[:,:-3]
mb_data_2_y = mb_data_2[:,-3:]

print(mb_data_2_x)
print(mb_data_2_y)

mb_data_3_x = mb_data_3[:,:-3]
mb_data_3_y = mb_data_3[:,-3:]

print(mb_data_3_x)
print(mb_data_3_y)

mb_data_4_x = mb_data_4[:,:-3]
mb_data_4_y = mb_data_4[:,-3:]
print(mb_data_4_x)
print(mb_data_4_y)
```
