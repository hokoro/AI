sigmoid 와 entropy 를 한번에 적용하여 구한식

```python
import numpy as np
def sigmoid(x):
  y = 1/(1+np.exp(-x))
  print(y)
  return y
```

```python
sigmoid(1000)
sigmoid(0)
```

```python
sigmoid(-10000) #음수 값이 뜨면서 경고를 띄운다 overflow 문제 
```

```python
def sigmoid_upgrade(x):
  y = np.exp(x)/(1+np.exp(x))
  print(y)
  return y
```

```python
sigmoid_upgrade(-10000) #양수가 커진 값은 처리 못함 
sigmoid_upgrade(0)
```

```python
def relu(x):
   return np.maximum(x,0) #두번째 기준에 따라서 0보다 크면 x 를 출력함  
  
```

```python
print(relu(150))
print(relu(0))
print(relu(-150))
```

```python
def sigmoid_upgrade2(x):
  y = np.exp(-relu(-x))/1+np.exp(-np.abs(x))
  print(y)
  return y
```

```python
x_list = [1000,0,-1000]
```

```python
for x in x_list:
  print(sigmoid(x))
```

```python
for x in x_list:
  print(sigmoid_upgrade(x))
```

```python
for x in x_list:
  print(sigmoid_upgrade2(x))
```

```python
#sigmoid cross entropy 

def sigmoid_cross_entropy(x,z):
  H = np.maximum(x,0) -(x *z) + np.log(1+np.exp(-np.abs(x)))
  return H 


```

```python
sigmoid_cross_entropy(1000,0.5)
```
