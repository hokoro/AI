```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0


model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', #loss = mse 를 해서 구할수도 있다.
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)
model.evaluate(x_test, y_test)
```

MSE = 평균 제곱 오차 
현재 신경망이 얼만만큼 차이가 나는지 정량적 수치를 나타내기 위해 

```python
#문제 Mse 를 구현 
import numpy as np
medal_predict = [15,12,18]
medal = [21,10,14]

medal_min = [a-b for a,b in zip(medal,medal_predict)]
medal_squr = [x**2 for x in medal_min]
medal_sum = sum(medal_squr)
medal_mean = medal_sum/len(medal)
#print("MSE : {}".format(medal_mean))
print(np.round(medal_mean,3))
```

##Xor 문제를 풀기 위해  다층 신경망 구조 로 풀기 
Dense 1개  = 1 layer
1개의 layer 에는 
노드=퍼셉트론 =유닛 이라는 값이 하나 이상 존재 한다.


```python
import tensorflow as tf 
import numpy as np 

x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]]) 
```

#신경망을 만들떄 일련의 순서 

##1. 데이터를 불러온다 
##2. 신경망 모델을 만든다.
##3. 최적화를 어떻게 할것인지 
##4. 학습 


```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(units = 2 , activation= 'elu',input_shape = (2,)),   # 모든 노드를 연결하는 방법 구조 : Dense(unit 의수, activation function,input_shape = ())                                  
  #tuple 은 값이 하나 들어올떄 (2,)
  tf.keras.layers.Dense(units = 5 , activation= 'elu'),
  tf.keras.layers.Dense(units = 7 , activation= 'elu'),
  tf.keras.layers.Dense(units = 1 , activation= 'sigmoid') # 0~1 사이의 값을 출력하기 위해 
] )
```

신경망 모델을 만드는 방법 2가지

1. 순서적으로
2. 원하는 방식대로 만드는 함수형 방식

최적화

learning rate = 학습률 학습에 정교함을 얼마나 정교하게 할지 수치를 만든다

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001),
              loss = 'mse')
history = model.fit(x,y,epochs = 1000) #순전파와 역전파를 모두 처리하는 역할이다.

```

```python
#예측 값 보기 
print(model.predict(x))
'''
[[0.02987769]
 [0.9752364 ]
 [0.9784889 ]
 [0.02612764]]
'''
```

one_hot vector
M  F   I 
1  0   0

0  0   1

단점 : 데이터의 수가 많으면 데이터가 영향을 줄정도로 잡아 먹는다.


```python
row1 = ['M',0.455,0.365,0.095]
row2 = ['F',0.455,0.365,0.095]
row3 = ['I',0.455,0.365,0.095]

result = np.zeros((3,3),dtype = np.int32)

if row1[0] == 'M':
  result[0][0] = 1
elif row1[0] == 'F':
  result[0][1] = 1
else:
  result[0][2] = 1
row1[0] = result[0]

if row2[0] == 'M':
  result[1][0] = 1
elif row2[0] == 'F':
  result[1][1] = 1
else:
  result[1][2] = 1
row2[0] = result[1]
if row3[0] == 'M':
  result[2][0] = 1
elif row3[0] == 'F':
  result[2][1] = 1
else:
  result[2][2] = 1
row3[0] = result[2]

print(result)
print(row1,'\n',row2,'\n',row3)

if row1[0] == 'M':
  row1[0] = [1,0,0]
elif row2[0] == 'F':
  row2[0] = [0,1,0]
else:
  row3[0] = [0,0,1]
print(row1,'\n',row2,'\n',row3)
```
