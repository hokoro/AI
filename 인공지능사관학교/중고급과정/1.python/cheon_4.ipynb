````python
#인구 500만 이하 
print(s[s<=5000000])
#인구 500만 이하 250만 이상
print(s[s[2500000<s] & s[s<=5000000]]) #표현에 모호성 때문에 and or 보다는 & ,| 이것을 pandas 에서 사용해야 한다.
print(s[s<=5000000][s>=2500000])
````

데이터 프레임 합치기

````python
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
print(df1)
print(df2)
````

````python
pd.concat([df1,df2],axis = 0,ignore_index=True) #리스트 형태로 합치는데 axis = 0 default 이기 때문에 행방향으로 합쳐진다.
#행방향 합치기 
````

````python
pd.concat([df1,df2],axis = 1,ignore_index=True) #열방향 합치기 
````

````python
pd.concat([df1,df2],axis = 1,ignore_index=True,join = 'inner') #교집합 조인 join = 'inner' 로 변경
````

데이터 프레임 병합

````python
#병합 merge
df1 = pd.read_excel(drive_path + 'stock price.xlsx')
df2 = pd.read_excel(drive_path + 'stock valuation.xlsx')

print(df1)
print(df2)
````

````python
#교집합
pd.merge(df1,df2) #기본적으로 교집합으로 합쳐진다.
pd.merge(df1,df2,how='outer',on = 'id') #on = id 로 통해 기준이 되는 id 도 병합을 한다. 단 데이터 없으면 None
#왼쪽 기준 키 값 분리 
pd.merge(df1,df2,how = 'left',left_on = 'stock_name',right_on = 'name')
#오른쪽 기준으로 분리 
pd.merge(df1,df2,how = 'right',left_on = 'stock_name',right_on = 'name')
````

누락데이터 치환

````python
#평균으로 누락데이터 치환 
df['age'].value_counts(dropna = False) #177개의 Nan 이 발생 
age_mean = df['age'].mean()
df['age'].fillna(age_mean,inplace = True) #평균값으로 Nan 을 채워라 

df['age'].value_counts(dropna = False)
````

````python
#빈도수로 누락데이터 치환 
embark_town_max = df['embark_town'].value_counts().idxmax() #가장 빈도가 많은 index 의 이름을 보여준다.
df['embark_town'].fillna(embark_town_max , inplace = True)
df.isnull().sum()
````

````python
#이웃하는 값으로 치환 
df['embarked'].fillna(method = 'ffill',inplace = True) #자기 앞에 있는 이웃으로 치환 ffill + 첫번쨰 데이터의 데이터가 있는지 확인해야 한다 핗수 
df.isnull().sum()
````

### 범주형 데이터 처리 (카테고리)


###구간 분할 
-연속 데이터를 그대로 사용하지 않고 일정 구간(bin) 으로 나눠서 분석 
-> 효율적인 경우 일때만 사용한다.

####원핫 인코딩 
-범주형 데이터를 0 과 1로만 구성된 것으로 변환 


###정규화 

-숫자 데이터 에 상대적인 크기 차이를 제거 한다.
데이터 가 0 ~ 500 까지 있는 데 
10~ 1000까지 있는데이터 랑 비교했을떄 영향을 많이 받는다 
따라서 두개의 데이터 크기를 비슷한 범위로 만들어 낸다.

보통 0~1 사이에 데이터로 바꾼다.

````python
df = sns.load_dataset('mpg')
df['horsepower']/abs(df['horsepower'].max()) #0~1 사이의 값으로 바꿔 정규화를 시켜준다.
#안통하는 데이터가 있다. 300 ~ 700 이라고 하면  max 700으로 모든 데이터를 나누면 원하는 값이 안나온다 따라서 300 ~ 700 데이터를 min  = 300 으로 뺴줘야 한다
df['horsepower'].min()
#뺴주지 안고 계산하면 이만큼에 오차가 발생한다.
````

````python
df['horsepower'] - df['horsepower'].min() / df['horsepower'].max()
````
