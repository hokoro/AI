```python

class Student:
  def __init__(self, name, age, grade):
    self.name = name
    self.age = age
    self.grade = grade
  
  def get_grade(self):
    return self.grade

class Course:
  def __init__(self, name, max_students):
    self.name = name
    self.max_students = max_students
    self.students = []

  def add_student(self, student):
    if self.max_students > len(self.students):
      self.students.append(student)
      return True
    else:
      print("정원 초과 되었습니다")
      return False
  def get_average_grade(self):
    sum = 0
    for student in self.students:
      sum += student.get_grade() #의존관계 클래스... 
    return sum / len(self.students)


s1 = Student('아이언맨', 19, 95)
s2 = Student('스파이더맨', 19, 75)
s3 = Student('헐그', 19, 85)

course = Course('과학', 2)
course.add_student(s1)
course.add_student(s2)
print(course.get_average_grade())
  

  
```

```python
def isDivisor(num1,num2):
  try:
    if num2 % num1 == 0:
      return True
    else:
      return False
  except ZeroDivisionError:
    print("0으로 못나눕니다")
  
num1 = int(input("첫번쨰 정수 입력: "))
num2 = int(input("두번째 정수 입력: "))

result = isDivisor(num1,num2)

print("결과: ",result)
```

```python
#getDivisor

def getDivisor(num):
  answer = []
  for i in range(1,num+1):
    if num % i ==0:
      answer.append(i)
  return answer


num = int(input())
result = getDivisor(num)
print(num,"의 약수:",end = '')  #10 의 약수: [1, 2, 5, 10]
for i in result:
  print(i , end = ' ')
```

class 상속

```python
#class 응용 상속 : 비슷한 두클래스를 줄이기 위해 중복방지를 위해 사용 
class Pet:
  def __init__(self,name,age):
      self.name = name
      self.age = age
  def show(self):
    print("나는 ",self.name,"이고",self.age,"세 입니다")
  def speak(self):
    print("없는 동물")
  

class Dog(Pet): #부모는 자식클래스 함수는 사용못함! but 자식은 부모 클래스 함수 사용가능
  def speak(self):
    print("멍")


class Cat(Pet):
  def speak(self):
    print("냐옹")

 
D = Dog('뽀빠',15)
C= Cat("베뤼",20)
P = Pet("팀",19)
D.show()
C.show()
P.show()
```

```python
class Fish(Pet):
  pass

f = Fish("버블",10)
f.speak()
f.show()
```

```python
class Cat(Pet):
  def __init__(self,name,age,color):
    super().__init__(name,age) #상위 클래스의 함수나 초기화 def 를 사용하기 위해서는 super를 사용해서 초기화를 할수 있다.
    self.color = color  
  def speak(self):
    print("냐옹")

c = Cat('빌',24,'brown')
c.name
```
