#### 그림 범위 지정

- 플롯 그림을 보면 몇몇 점들은 그림의 범위 경계선에 있어서 잘 보이지 않는 경우
- 그림의 범위를 수동으로 지정하려면 xlim 명령과 ylim 명령을 사용
- 그림의 범위가 되는 x축, y축의 최소값과 최대값을 지정


````python
plt.xlim(-2,8)#xlim, ylim = x,y의 화면에 보여질 범위를 지정해 준다.
plt.ylim(-5,10)
plt.plot(x,y,ls = '--',marker = 'o',mec = 'g',ms ='5',mew = '1',mfc = 'g')
plt.show()
````

#### Matplotlib 연습 02-2

교통사고 다발 시간대 시각화

> 0-2, 3-5, 6-8, 9-11, 12-14, 15-17, 18-20, 21-23시 8개의 구간으로 나누어서 진행


````python
time_df = df['발생년월일시'] % 100
time_df.value_counts().sort_index()
print(time_df.values)

bins = [-1, 2, 5, 8, 11, 14, 17, 20, 24]
labels = ['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18-20', '21-23']
time_cuts = pd.cut(time_df.values, bins, labels=labels)
rs = time_cuts.value_counts() #전체 데이터를 구간별로 카운팅
print(rs)
#print(rs[['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18-20', '21-23']])
x = rs.index #rs 의 index 추출
y = rs.values #rs 의 값 추출
plt.bar(x, y) #바 데이터 만들기 
plt.show()
````

````python
#파이 차트 
plt.figure(figsize=(8,8)) #plot 차트의 크기를 구하는 식
plt.subplot(2,2,1)
plt.pie(rs,labels = labels,autopct='%1.1f%%') #파이차트(data,label,autopct = 몇%인지)
plt.subplot(2,2,2)
plt.pie(rs,labels = labels,autopct='%1.1f%%')
plt.subplot(2,2,4) #subplot(x,y,내가 넣을 plot 위치)
plt.pie(rs,labels = labels,autopct='%1.1f%%') 
plt.show() 
````

````python
# 파이차트
labels = ('A', 'B', 'C') 
xs = [50, 30, 40] 

patches, texts, autotexts = plt.pie(
    labels=labels, 
    labeldistance=1.1,   # label과 파이의 간격, 1일 경우 딱 붙음. 
    x = xs,              # 값
    explode=(0.1, 0.1, 0), # 파이조각 돌출 크기 (0 이면 돌출되지 않음)
    startangle=90,       # 그려지는 파이 조각 시작위치. 90은 12시 방향 
    shadow=False,        # 파이차트 그림자 효과 유무
    counterclock=False,  # 파이차트가 그려질 방향 (시계방향, 반시계 방향) 
    autopct='%1.1f%%',   # 파이조각 위에 표시될 글자 형태 (파이조각 전체대비 백분율)
    pctdistance=0.7,     # 백분률 표시 위치 (radius 기준으로 위치 선정)
    colors=['grey', 'red', 'blue'],  # 각 파이조각 컬러
)
````
